##### Q. What difference do you notice in the size of the 3rd dimension across .jpg and .png images?
> The .jpg file's 3rd dimension is of size 3. The .png file's 3rd dimension is of size 4. This is because the .png image has an alpha channel and the .jpg image doesn't. The alpha channel controls the level of opacity of the image.

### Task 13

```python
gamma = 1 / 5
new_png = pineapple_png[:, :, 0:3] ** gamma
plt.imshow(new_png)
```

```python
gamma = 5
new_png2 = pineapple_png[:, :, 0:3] ** gamma
plt.imshow(new_png2)
```

```python
gamma = 1
new_png3 = pineapple_png[:, :, 0:3] ** gamma
plt.imshow(new_png3)
```

```python
fig, ax = plt.subplots(1, 3)
fig.suptitle('Horizantally Stacked Subplots')
ax[0].set_title('Gamma = 1 / 5')
ax[1].set_title('Gamma = 5')
ax[2].set_title('Gamma = 1')
ax[0].set_yticks([])
ax[1].set_yticks([])
ax[2].set_yticks([])
ax[0].imshow(new_png)
ax[1].imshow(new_png2)
ax[2].imshow(new_png3)
```
